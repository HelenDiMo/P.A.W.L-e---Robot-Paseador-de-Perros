```python
import random
import time

def pasear_perro():
    print("--- 🕒 Inicio del paseo ---")
    print("1. Localizando al perro...")
    print("2. Poniendo la correa y abriendo la puerta.")
    print("3. Saliendo de casa y cerrando con llave.")

    tiempo_transcurrido = 0
    objetivo_tiempo = 30  # minutos
    esta_lloviendo = False

    # Bucle principal del paseo
    while tiempo_transcurrido < objetivo_tiempo:
        print(f"\n[Minuto {tiempo_transcurrido}] Paseando...")

        # Simulación de eventos aleatorios
        evento = random.randint(1, 10) # Genera un número del 1 al 10

        # Reto de la lluvia: Si el número es 10, empieza a llover
        if evento == 10:
            print("⚠️ ¡ALERTA! Ha empezado a llover fuertemente.")
            esta_lloviendo = True
            break # El break rompe el bucle y sale del paseo inmediatamente

        # Simulación de obstáculos (charcos)
        elif evento == 1 or evento == 2:
            print("💧 Charco detectado. Desplazándose a la derecha para esquivar.")

        # Simulación de necesidades del perro
        elif evento == 5:
            print("💩 El perro se ha detenido. Recogiendo excremento con la pinza...")
            print("♻️ Bolsa depositada en la papelera más cercana.")

        # Simulamos que pasa el tiempo (1 segundo real = 5 minutos de paseo)
        time.sleep(1)
        tiempo_transcurrido += 5

    # Finalización del paseo
    if esta_lloviendo:
        print("\n🏠 Regresando a casa rápido por la lluvia...")
    else:
        print(f"\n✅ Tiempo cumplido ({tiempo_transcurrido} min). Regresando a casa.")

    print("4. Entrando en casa y quitando la correa.")
    print("🤖 Robot: 'He completado mi tarea con éxito'.")
    print("🔌 Volviendo a la base de carga.")

# Ejecutamos la función
pasear_perro()
```

```python
import random
import time

def menu_inicio():
    # Preguntamos al usuario y guardamos la respuesta en la variable 'confirmacion'
    # .lower() sirve para que acepte "SI", "Si" o "si" por igual
    confirmacion = input("¿Deseas comenzar el paseo del perro ahora? (si/no): ").lower()

    if confirmacion == "si":
        print("\n🤖 Robot: 'Entendido. Iniciando protocolo de paseo...'")
        time.sleep(1) # Una pequeña pausa para dar realismo
        pasear_perro()
    else:
        print("\n🤖 Robot: 'De acuerdo. Me quedaré en la base de carga hasta que me necesites.'")

def pasear_perro():
    print("\n--- 🕒 Fase 1: Preparación ---")
    print("1. Localizando al perro en la casa...")
    time.sleep(1)
    print("2. Coger la correa del bolsillo interior.")
    print("3. Poner la correa al perro.")
    print("4. Abrir la puerta, salir y cerrar con llave.")

    tiempo_transcurrido = 0
    objetivo_tiempo = 30
    esta_lloviendo = False

    # Bucle del paseo
    while tiempo_transcurrido < objetivo_tiempo:
        print(f"\n[Minuto {tiempo_transcurrido}] Paseando...")
        evento = random.randint(1, 10)

        if evento == 10:
            print("⚠️ ¡ALERTA! Ha empezado a llover.")
            esta_lloviendo = True
            break

        elif evento <= 2:
            print("💧 Esquivando un charco.")

        elif evento == 5:
            print("💩 Recogiendo excremento y depositándolo en la basura.")

        time.sleep(1)
        tiempo_transcurrido += 5

    # --- RESPUESTA AL RETO ANTERIOR ---
    if esta_lloviendo:
        print("\n🏠 Regreso precipitado a casa.")
        print("🤖 Robot: 'Tarea interrumpida por el clima'.")
    else:
        print(f"\n✅ Paseo completado con éxito.")
        print("🤖 Robot: 'He completado mi tarea con éxito'.")

    print("5. Entrar, quitar correa y volver a la base de carga.")

# Ahora llamamos primero al menú
menu_inicio()
```
